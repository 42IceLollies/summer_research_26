```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.lines as mlines

target_size = 6
```

```python
# Dot Grid
fig, ax = plt.subplots(figsize = (target_size, target_size))

circ_size = 0.001
circ_spacing = 0.01
num_circs = int(1/(circ_spacing + 2*circ_size))

for i in range(num_circs):
    for j in range(num_circs):
        x = 2*circ_size*i+circ_spacing*i+circ_spacing
        y = 2*circ_size*j+circ_spacing*j+circ_spacing
        circ = mpatches.Circle((x,y), circ_size, color = "black")
        ax.add_patch(circ)


plt.show()
```

```python
# Windowpane Grid
fig, ax = plt.subplots(figsize = (target_size, target_size))

line_spacing = 0.015
num_lines = int(1/line_spacing)

for i in range(num_lines):
    x = line_spacing *i + line_spacing
    line_y = mlines.Line2D((x,x), (0,1), linewidth = 0.5, color = "black")
    line_x = mlines.Line2D((0,1), (x,x), linewidth = 0.5, color = "black")
    ax.add_line(line_y)
    ax.add_line(line_x)
plt.show()
```

```python
# Edgy one
fig, ax = plt.subplots(figsize = (target_size, target_size))

line_spacing = 0.03
num_lines = int(1/(2*line_spacing))
print(num_lines)

for i in range(num_lines):
    x = 2*line_spacing*i+line_spacing
    rect = mpatches.Rectangle((x,0), line_spacing, 1, color = "black")
    ax.add_patch(rect)

plt.show()
```

```python
# checkerboard
fig, ax = plt.subplots(figsize = (target_size, target_size))

line_spacing = 0.03
num_lines = int(1/(line_spacing))
print(num_lines)

for i in range(num_lines):
    for j in range(num_lines):
        x = line_spacing*i
        y = line_spacing*j
        if i%2 == j%2:
            rect = mpatches.Rectangle((x,y), line_spacing, line_spacing, color = "black")
        else: 
            rect = mpatches.Rectangle((x,y), line_spacing, line_spacing, color = "white")
        ax.add_patch(rect)

plt.show()
```
